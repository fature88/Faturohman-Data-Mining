```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("laptop_price.csv")
```

```python
# Ukuran dan info dataset
print(df.shape)
print(df.info())
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Price (euros)'], bins=30, kde=True)
plt.title('Distribusi Harga Laptop')
plt.xlabel('Harga (Euro)')
plt.ylabel('Jumlah')
plt.show()
```

```python
print("Sebelum dibersihkan:", df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()             # Hilangkan spasi depan/belakang
df.columns = df.columns.str.replace('\r', '')   # Hilangkan karakter carriage return
df.columns = df.columns.str.replace('\n', '')
```

```python
print("Setelah dibersihkan:", df.columns.tolist())
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Price (euros)'], bins=30, kde=True)
plt.title('Distribusi Harga Laptop')
plt.xlabel('Harga (Euro)')
plt.ylabel('Jumlah')
plt.show()
```

```python
'Price (Euro)'
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Price (Euro)'], bins=30, kde=True)
plt.title('Distribusi Harga Laptop')
plt.xlabel('Harga (Euro)')
plt.ylabel('Jumlah')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='TypeName', y='Price (Euro)', estimator='mean')
plt.title('Rata-rata Harga per Tipe Laptop')
plt.xticks(rotation=45)
plt.ylabel('Rata-rata Harga (Euro)')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='OpSys', y='Price (Euro)')
plt.title('Harga Laptop berdasarkan Sistem Operasi')
plt.xticks(rotation=45)
plt.ylabel('Harga (Euro)')
plt.show()
```

```python
plt.figure(figsize=(14,6))
top_companies = df['Company'].value_counts().index
sns.barplot(data=df, x='Company', y='Price (Euro)', estimator=np.mean, order=top_companies)
plt.title('Rata-rata Harga Laptop per Perusahaan')
plt.xticks(rotation=45)
plt.ylabel('Rata-rata Harga (Euro)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='RAM (GB)', y='Price (Euro)')
plt.title('Harga Laptop Berdasarkan RAM')
plt.xlabel('RAM (GB)')
plt.ylabel('Harga (Euro)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Inches', y='Price (Euro)', hue='Company')
plt.title('Ukuran Layar vs Harga Laptop')
plt.xlabel('Ukuran Layar (Inch)')
plt.ylabel('Harga (Euro)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

```python
plt.figure(figsize=(14,6))
top_gpu_types = df['GPU_Type'].value_counts().index[:15]
sns.countplot(data=df[df['GPU_Type'].isin(top_gpu_types)], x='GPU_Type', order=top_gpu_types)
plt.title('Jumlah Laptop Berdasarkan Jenis GPU (Top 15)')
plt.xticks(rotation=45)
plt.xlabel('GPU Type')
plt.ylabel('Jumlah')
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Fitur Numerik')
plt.show()
```
